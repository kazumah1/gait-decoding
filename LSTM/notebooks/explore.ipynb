```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
train_df = pd.read_csv("../gait-data/train_set.csv")
labels = pd.read_csv("../gait-data/labels_train_set.csv")
merged_df = pd.merge(train_df, labels, on="ID")
merged_df.sample(10)
```

```python
merged_df = merged_df.sort_values(by=["Subject", "Session", "Time:512Hz"])
```

```python
subject1_session1 = merged_df[(merged_df['Subject'] == "SCI2") & (merged_df['Session'] == 1)]
```

```python
subject1_session1
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=subject1_session1['time'].head(5000),
        y=subject1_session1['angle'].head(5000),
        mode='lines',
        name='Angle',
    )
)
fig.update_layout(
    title='Foot Angle Over Time (from first 100 rows of labels_df)',
    xaxis_title='Time (s)',
    yaxis_title='Angle (radians)',
    width=1200,
    height=600,
)
fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)
fig.show()
```
